```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

path='H:/Thesis/[4-2]/Dataset/train/WorkingDataset/4-PDR/3993_right.jpeg'
image = cv2.imread(path)
gray_image=cv2.imread(path)
'''Convert in gray'''
#gray_image=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

''' Cropping'''
h,w=gray_image.shape[:2]
x1,y1=int(h*.05),int(w*.15)
x2,y2=int(h*.95),int(w*.85)
croped_image=gray_image[x1:x2,y1:y2]
cv2.imshow('After cropping ',croped_image)

'''Resizing'''
i_s3=cv2.resize(croped_image,(612,512),interpolation=cv2.INTER_AREA)
cv2.imshow('3.Scaling- Skewed Interpolation',i_s3)
b,g,r = cv2.split(i_s3)
cv2.imshow('B ',b)
cv2.imshow('G ',g)
cv2.imshow('R ',r)


histogram=cv2.calcHist([b],[0],None,[256],[0,256])
plt.hist(b.ravel(),256,[0,256])
plt.show()

histogram=cv2.calcHist([g],[0],None,[256],[0,256])
plt.hist(g.ravel(),256,[0,256])
plt.show()

histogram=cv2.calcHist([r],[0],None,[256],[0,256])
plt.hist(r.ravel(),256,[0,256])
plt.show()

equ = cv2.equalizeHist(g)

histogram=cv2.calcHist([equ],[0],None,[256],[0,256])
plt.hist(equ.ravel(),256,[0,256])
plt.show()

cv2.imshow('Equ ',equ)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
